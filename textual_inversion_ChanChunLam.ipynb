```python
!python main.py --base configs/latent-diffusion/txt2img-1p4B-finetune_vpt2.yaml -t --no-test --actual_resume models/ldm/text2img-large/model.ckpt -n recreate --gpus 0, --data_root datasets/cat_statue --init_word cat
!python main.py --base configs/latent-diffusion/txt2img-1p4B-finetune_vpt4.yaml -t --no-test --actual_resume models/ldm/text2img-large/model.ckpt -n recreate --gpus 0, --data_root datasets/cat_statue --init_word cat

#extracted val/loss of the above from their logs/corresponding_name/testtube/version_0/metrics.csv into cat1.csv, cat2.csv and cat4.csv

import pandas as pd
import matplotlib.pyplot as plt
cat1 = pd.read_csv("plot/cat1.csv")
cat1 = cat1["val/loss"]
cat2 = pd.read_csv("plot/cat2.csv")
cat2 = cat2["val/loss"]
cat4 = pd.read_csv("plot/cat4.csv")
cat4 = cat4["val/loss"]
cat = pd.DataFrame(data={'cat1':cat1, 'cat2': cat2, 'cat4': cat4})
cat = cat.dropna(axis = 0, how = 'all')
cat["cat1"]=cat["cat1"].fillna(method='ffill')
cat["cat2"]=cat["cat2"].fillna(method='ffill')
cat["cat4"]=cat["cat4"].fillna(method='ffill')
plt.plot(cat["cat1"], label = "1 vector")
plt.plot(cat["cat2"], label = "2 vectors")
plt.plot(cat["cat4"], label = "4 vectors")
plt.legend()
plt.title("val/loss for different numbers of vectors per token")
plt.ylabel("val/loss")
plt.show()
```
